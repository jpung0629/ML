```python
# 생활코딩 코딩야학 머신 러닝 1 레모네이드 예측하기 입니다.
# 라이브러리 임포트
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
# 1. 데이터 받아오기
# 2. 종속 변수, 독립 변수 분리
# 레모네이드
# 온도, 판매량
# 독립 변수: 온도
# 종속 변수: 판매량
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'

frame = pd.read_csv(path)

independent_values = frame[['온도']]
dependent_values = frame[['판매량']]

print(f'i: {independent_values}')
print(f'd: {dependent_values}')
```

```python
# 모델 생성
X = tf.keras.layers.Input(shape=[1])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
model.fit(independent_values, dependent_values, epochs=30000)
```

```python
print(model.predict(independent_values))
print(dependent_values)
```
