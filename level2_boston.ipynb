```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
# 데이터 준비
path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'

boston = pd.read_csv(path)

i = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm','age','dis','rad','tax','ptratio','b','lstat']]
d = boston[['medv']]

print(f'i: {i.shape}')
print(f'd: {d.shape}')
```

```python
# 모델 구조 생성
# 입력층
X = tf.keras.layers.Input(shape=[13])

# 출력층
Y = tf.keras.layers.Dense(1)(X)

# 모델
model = tf.keras.models.Model(X, Y)

# 손실 함수는 평균 제곱 오차 함수
model.compile(loss='mse')
```

```python
# 학습(fit) 한다
model.fit(i, d, epochs=1000)
```

```python
# 모델 사용
# 예측치 출력
print(model.predict(i[1:10]))

# 실제 값 출력
print(d[1:10])
```

```python
# 추가 학습 후 예측
model.fit(i, d, epochs=1000)
print(model.predict(i[1:10]))
print(d[1:10])
```

```python
# loss 값이 24 이하로 떨어지지 않기 때문에 학습 종료.
# 학습 후의 가중치 값 출력
model.get_weights()
```
